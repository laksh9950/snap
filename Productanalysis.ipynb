```python
def cated(product):
    if 'Sugar' in product.split():
        return "sugar"
    
    elif 'Oil' in product.split():
        return "oil"
    elif 'Dal' in product.split():
        return "Dal"
    
```

```python
df['Product Name'].apply(cated)
```
